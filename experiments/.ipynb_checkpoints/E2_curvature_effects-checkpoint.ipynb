```python
import os
import torch
import matplotlib.pyplot as plt
from datetime import datetime

from src.diffeomorphisms.vanilla_hyperboloid import Vanilla_into_Hyperboloid
from src.manifolds.pull_back_manifold import PullBackManifold
from src.riemannian_autoencoder import Curvature_Corrected_Riemannian_Autoencoder
from src.riemannian_autoencoder.low_rank_approximation.naive_tsvd import naive_low_rank_approximation

# set seed
torch.manual_seed(31)
```

```python
# Construct spiral data
def hyperbolic(n_points, noise=0.1): # returns a data set that are geodesics in hyperbolic space -- compute geodesic in poincare disc and map out
    """
     Returns the two spirals dataset.
    """
    n = torch.linspace(-2.,2.,n_points)[:,None]
    d1x = torch.sinh(n)
    d1y = torch.cosh(n)
    return torch.hstack((d1x,d1y)) + torch.randn((n_points,2)) * noise

num_data = 51
data = hyperbolic(num_data, noise=0.1) 

plt.title('data set')
plt.plot(data[:,0], data[:,1], '.')
plt.show()
```

```python
# construct pull-back manifold
offset = torch.zeros((2,))
orthogonal = torch.eye(2,2)
diffeo = Vanilla_into_Hyperboloid(2, offset, orthogonal)

M = PullBackManifold(diffeo)
```

Evaluate learned geometry

```python
# choose special points
z1 = data[0][None]
z2 = data[int(num_data/2)][None]
z3 = data[int(3 *num_data/4)][None]
z4 = torch.tensor([0., 1.])[None]
```

```python
# compute geodesics
t = torch.linspace(0,1,100)
gamma_12_t =  M.geodesic(z1, z2, t)
gamma_13_t =  M.geodesic(z1, z3, t)
```

```python
# plot geodesics
plt.plot(data[:,0], data[:,1], '.')
plt.plot(gamma_12_t.squeeze().detach().numpy()[:,0], gamma_12_t.squeeze().detach().numpy()[:,1], '.')
plt.plot(gamma_13_t.squeeze().detach().numpy()[:,0], gamma_13_t.squeeze().detach().numpy()[:,1], '.')
plt.show()
```

```python
# compute barycentre
barycentre = M.barycentre(data[None])
```

```python
# plot barycentre
plt.plot(data[:,0], data[:,1], '.')
plt.plot(barycentre[:,0].detach().numpy(), barycentre[:,1].detach().numpy(), '.')
plt.show()
```

```python
# compute logs to all data
logs = M.log(barycentre, data[None])[0]
```

```python
plt.plot(logs.squeeze().detach().numpy()[:,0], logs.squeeze().detach().numpy()[:,1], '.')
plt.axis('equal')
plt.show()
```

```python
# rank 1 approximation
z = barycentre.squeeze()
R_z, U = naive_low_rank_approximation(M, z, data, 1)
w_z = R_z / M.norm(z[None], R_z[None])[:,None]
print(w_z)
print(U.shape)
```

```python
# compute error
logs_approx = torch.einsum("kN,kd->Nd", U, R_z)

tangent_space_error = torch.sum(M.norm(z[None],logs_approx[None] - logs[None])**2)
print(tangent_space_error)

manifold_error = torch.sum(M.distance(data[:,None], M.exp(z[None],logs_approx[None])[0][:,None])**2)
print(manifold_error)

# TODO implement curvature corrected low rank approximation
```

```python
rae_w_z = Curvature_Corrected_Riemannian_Autoencoder(M, z, w_z)
```

```python
# project noisy data onto manifold with RAE
noisy_data = hyperbolic(500, noise=0.1) 
projected_mesh = rae_w_z.project_on_manifold(noisy_data)
```

```python
# plot results
plt.plot(noisy_data[:,0], noisy_data[:,1], '.')
plt.plot(projected_mesh.cpu().detach().numpy()[:,0], projected_mesh.cpu().detach().numpy()[:,1], '.')
plt.show()
```
